```python
# 3.2 Convert Money-Related Columns to Numeric
money_cols = [col for col in df.columns if ("net_assets" in col) or col.endswith("_asset")]
print("Money-related columns to convert:", money_cols)

# Loop over each money column and format it
for col in money_cols:
    df[col] = df[col].replace({'US\$': '', ',': ''}, regex=True)
    df[col] = pd.to_numeric(df[col], errors='coerce')
    # Show the first few values after conversion for confirmation
    print(f"Column '{col}' after conversion:")
    print(df[col].head(), "\n")
```

```python
print(df.columns.tolist())
```

```python
# 4. ESG Fund Filtering
df['is_esg'] = (df['sustainability_mandate'].fillna('') == 'Y') | (df['us_sif_member'].fillna('') == 'Y')
print("Number of ESG funds:", df['is_esg'].sum())
df_esg = df[df['is_esg']]
df_esg[['fund_name', 'sustainability_mandate', 'us_sif_member']].head()
```

```python
# 6. Analysis by Fund Category
esg_by_category = df.groupby("category")['is_esg'].sum().sort_values(ascending=False)
print(esg_by_category)
```

```python
plt.figure(figsize=(10,6))
esg_by_category.plot(kind='bar')
plt.title("Number of ESG Funds by Category")
plt.xlabel("Fund Category")
plt.ylabel("ESG Fund Count")
plt.tight_layout()
plt.show()
```

```python
# 7. Net Assets Distribution
plt.figure(figsize=(10,6))
plt.hist(df['net_assets'].dropna(), bins=30)
plt.title("Distribution of Fund Net Assets")
plt.xlabel("Net Assets (USD)")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

```python
# 8. Fund Inception Year Trend Analysis
df['inception_date'] = pd.to_datetime(df['inception_date'], errors='coerce')
df['inception_year'] = df['inception_date'].dt.year
inception_counts = df['inception_year'].value_counts().sort_index()
print(inception_counts)
```

```python
plt.figure(figsize=(10,6))
inception_counts.plot(kind='bar')
plt.title("Number of Funds Established per Year")
plt.xlabel("Year")
plt.ylabel("Fund Count")
plt.tight_layout()
plt.show()
```

```python
# 9. Fossil Fuel Exposure Analysis Example
if "fossil_holdings_weight" in df.columns:
    df['fossil_holdings_weight'] = pd.to_numeric(df["fossil_holdings_weight"], errors='coerce')
    plt.figure(figsize=(10,6))
    plt.hist(df['fossil_holdings_weight'].dropna(), bins=30)
    plt.title("Distribution of Fossil Fuel Holdings Weight")
    plt.xlabel("Holdings Weight (%)")
    plt.ylabel("Frequency")
    plt.tight_layout()
    plt.show()
else:
    print("The dataset does not contain the column 'fossil_holdings_weight'.")
```
